```python
df = pd.read_csv('../data/adri_data/entrenamiento_postulaciones_vistas18_full.csv', encoding='utf-8')
df.head()
```

```python
df['se_postulo'] = df['se_postulo'].apply(lambda x: 1 if x > 0 else 0)
df['se_postulo'].unique()
```

```python
df.drop(['titulo', 'descripcion', 'contenido'], inplace=True, axis=1)
```

```python
df['decil'] = (df['edad'] // 10) * 10
```

```python
def dropCols(df):
    cols = df.head(0)    
    cols.drop(['idaviso', 'idpostulante', 'se_postulo'], axis=1, inplace=True)    
    
    # cols.drop(['titulo', 'descripcion', 'contenido'
    #           # 'sepostulo'
    #           ], axis=1, inplace=True)    
    
    # cols = cols[['idtipo_de_trabajo_int', 'idnivel_laboral_int', 'idnombre_area_int',
    #    'month', 'day', 'dayofweek', 'hour', 
    #              'sexo', 'edad', 
    #              # 't_0', 'd_0','c_0', 
    #              # 't_1', 'd_1', 'c_1', 
    #              # 't_2', 'd_2', 'c_2', 
    #              # 't_3', 'd_3', 'c_3'
    #             ]]
    return cols.columns

cols = dropCols(df)
cols
```

```python
dfix = df.loc[:,cols];
dfix.head()
```

```python
# X will have the numeric data, Y will have the target class

X = np.array(dfix)
y = np.array(df['se_postulo']) 
```

```python
del dfix
```

```python
clf = make_pipeline(preprocessing.QuantileTransformer(), # PCA(n_components=3), 
                     DecisionTreeClassifier(max_depth=25))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7506)
clf.fit(X_train, y_train)   

clf.score(X_test, y_test)

pred = clf.predict(X_test)

print (cols)
print (accuracy_score(y_test, pred))
```

```python
# X will have the numeric data, Y will have the target class
dfix = df.loc[:,cols]
X = np.array(dfix)
y = np.array(df['se_postulo']) 
```

```python
# pca = PCA(n_components=3)
# pca.fit(X)
# X = pca.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=7506)
X_test
```

```python
y_test
```

```python
for i in range(11, 12):    
    knn = KNeighborsClassifier(n_neighbors=i)                 # Create KNN classifier
    knn.fit(X_train, y_train)                                 # FIT (this does nothing in normal KNN)
    pred = knn.predict(X_test)                                # predict the response
    print (str(i) + ": " + str(accuracy_score(y_test, pred))) # evaluate accuracy

# Con PCA
    
# 01: 0.853781296182839
# 02: 0.8552299723041294
# 03: 0.8617791956024632
# 04: 0.8622319068903664
# 05: 0.864944150517449
# 06: 0.8650709096780619
# 07: 0.8663274350082645
# 08: 0.8660648624612806
# 09: 0.8666674714867341
# 10: 0.8665205918244365
# 11: 0.8669541886357395
# 12: 0.8666141521572699
# 13: 0.8668686765035799
# 14: 0.8665779352542377
# 15: 0.8666211943328594
# 16: 0.8663606338360441
# 17: 0.8662590253025368
# 18: 0.8661232119161659
# 19: 0.8663314591086014

# Sin PCA

# 1: 0.8411124222971376
```

# Set de Entrenamiento de Postulaciones  

```python
avisos_detalles = pd.read_csv('../data/adri_data/avisos_detalles.csv')
avisos_detalles.head()
```

```python
postulaciones = pd.read_csv('../data/adri_data/postulaciones.csv')
postulaciones.head()
```

```python
postulaciones_sin_descripcion = pd.merge(avisos_detalles, postulaciones, how='inner', on='idaviso')
postulaciones_sin_descripcion.head()
```

```python
postulantes = pd.read_csv('../data/adri_data/postulantes2.csv')
postulantes.head()
```

```python
postulaciones_sin_descripcion = pd.merge(postulaciones_sin_descripcion, postulantes, how='inner', on='idpostulante')
postulaciones_sin_descripcion.count()
```

```python
postulaciones_sin_descripcion.to_csv('../data/adri_data/entrenamiento_postulaciones.csv', sep=',', encoding='utf-8', index=False)
postulaciones_sin_descripcion.head()
```

```python
vistas.drop('timestamp', axis=1, inplace=True)
```

```python
#vistas_no_post = vistas.merge(postulaciones, how='left', on=['idaviso','idpostulante'])
common = vistas.merge(postulaciones, on=['idaviso','idpostulante'])
no_pub = vistas[(~vistas.idaviso.isin(common.idaviso))&(~vistas.idpostulante.isin(common.idpostulante))]
```

```python
no_pub.head(20)
```

```python
no_pub['se_postulo'] = 0
no_pub.head()
```

```python
postulaciones['se_postulo'] = 1
postulaciones.head()
```

```python
postulaciones = postulaciones[['idaviso', 'idpostulante', 'valor_educacion', 'sexo', 'edad', 'idtitulo', 'idtipo_de_trabajo', 'idnivel_laboral', 'idnombre_area', 'iddenominacion_empresa', 'se_postulo']]
postulaciones.head()
```

```python
frames = [postulaciones, no_pub]
full_data = pd.concat(frames)
full_data.head()
```

```python
full_data = full_data.sample(frac=1).reset_index(drop=True)
full_data
```
